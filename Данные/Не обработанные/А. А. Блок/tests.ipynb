```python
from natasha_func import get_doc
from natasha_func import get_raw_text
def get_lemma_part_speech(doc):
    lemma_part_speech = {}
    for token in doc.tokens:
        if token.pos not in lemma_part_speech.keys():
            lemma_part_speech[token.pos] = []
        else:
            lemma_part_speech[token.pos].append(token.lemma)
    return lemma_part_speech

get_lemma_part_speech(get_doc(get_raw_text('blok_a_a-text_1906_devushka.fb2')))
```
